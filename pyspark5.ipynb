```python
import pyspark
```

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('practise').getOrCreate()
```

```python
df_pyspark=spark.read.csv('data4.csv', header=True, inferSchema=True)
```

```python
df_pyspark.show()
```

```python
### groupby

df_pyspark.groupBy('department').sum().show()
```

```python

df_pyspark.groupBy('department').count().show()
```

```python
df_pyspark.agg({'salary':'sum'}).show()
```

```python
from pyspark.sql.functions import*
#for using agg func like max, min, col
df_pyspark.select(max('salary')).show()
```

```python
a=df_pyspark.select('Name', 'salary').orderBy('salary', ascending=False).first()
#max_salary_row = df_pyspark.select('Name', 'salary').orderBy(col('salary').desc()).first()
```
