```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('practise').getOrCreate()
```

```python
df_pyspark=spark.read.csv('data2.csv', header=True, inferSchema=True)
```

```python
##drop the column
df_pyspark.drop("Name").show()
```

```python
df_pyspark.show()
```

```python
## removing null values
df_pyspark.na.drop().show()
```

```python
df_pyspark.na.drop(how="all").show()
# the row get dropped if all coumn has null value
```

```python
df_pyspark.na.drop(how="any").show()
```

```python
df_pyspark.na.drop(how="any", thresh=2).show()
# minimum 2 null values must be present
```

```python
df_pyspark.na.drop(how="any", subset=['age']).show()
#null value in particular cloumn
```

```python
###filling the missing the value
df_pyspark.na.fill(value=0).show()
```

```python
df_pyspark.na.fill(0,['experience', 'age']).show()
```

```python
df_pyspark.show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['age', 'experience', 'salary'],
    outputCols=["{}_imputed".format(c) for c in ['age', 'experience', 'salary']]
).setStrategy("mean") #median #mode
```

```python
#add imputation cols to df
imputer.fit(df_pyspark).transform(df_pyspark).show()
```
