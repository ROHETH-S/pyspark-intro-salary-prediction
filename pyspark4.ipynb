```python
import pyspark
```

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName('practise').getOrCreate()
```

```python
df_pyspark=spark.read.csv('data3.csv', header=True, inferSchema=True)
```

```python
df_pyspark.show()
```

Filter operation

```python
df_pyspark.filter("salary<=20000").show()
```

```python
df_pyspark.filter("salary<=20000").select(['Name', 'age']).show()
```

```python
df_pyspark.filter((df_pyspark['salary']<=50000) & (df_pyspark['salary']>=20000)).show()
```

```python
df_pyspark.filter(~(df_pyspark['salary']<=20000)).show()
```

```python
df_pyspark.filter(df_pyspark['age'] == 21).show()
```

```python
df_pyspark.filter((df_pyspark['age'] == 21) | (df_pyspark['age'] == 25)).show()
```
